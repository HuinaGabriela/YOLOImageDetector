```python
# 5. Dividir dataset em treino e validação

import os
import shutil
import random

def split_dataset(src_img_dir, src_label_dir, train_img_dir, val_img_dir, train_label_dir, val_label_dir, val_ratio=0.2):
    os.makedirs(train_img_dir, exist_ok=True)
    os.makedirs(val_img_dir, exist_ok=True)
    os.makedirs(train_label_dir, exist_ok=True)
    os.makedirs(val_label_dir, exist_ok=True)

    imgs = [f for f in os.listdir(src_img_dir) if f.endswith(('.jpg', '.png'))]
    random.shuffle(imgs)

    n_val = int(len(imgs) * val_ratio)
    val_imgs = imgs[:n_val]
    train_imgs = imgs[n_val:]

    for img in train_imgs:
        shutil.copy(os.path.join(src_img_dir, img), train_img_dir)
        label_file = img.rsplit('.', 1)[0] + '.txt'
        shutil.copy(os.path.join(src_label_dir, label_file), train_label_dir)

    for img in val_imgs:
        shutil.copy(os.path.join(src_img_dir, img), val_img_dir)
        label_file = img.rsplit('.', 1)[0] + '.txt'
        shutil.copy(os.path.join(src_label_dir, label_file), val_label_dir)

# Exemplo de uso:
split_dataset('/content/dataset/images', '/content/dataset/labels',
              '/content/dataset/images/train', '/content/dataset/images/val',
              '/content/dataset/labels/train', '/content/dataset/labels/val')
```

```python
# Ajuste proporcional: 80% treino, 20% validação
from sklearn.model_selection import train_test_split
import os
import shutil
from glob import glob

# Caminhos
image_dir = '/content/dataset/datasets/images'
label_dir = '/content/dataset/datasets/labels'

# Listar imagens
images = glob(os.path.join(image_dir, '*.jpg'))
images.sort()

# Separar treino e validação
train_imgs, val_imgs = train_test_split(images, test_size=0.2, random_state=42)

# Função para mover
def mover(lista, destino_imgs, destino_lbls):
    for img_path in lista:
        # nome base da imagem
        base = os.path.basename(img_path).replace('.jpg', '')
        label_path = os.path.join(label_dir, base + '.txt')

        # destino
        shutil.copy(img_path, destino_imgs)
        if os.path.exists(label_path):
            shutil.copy(label_path, destino_lbls)

# Mover arquivos
mover(train_imgs, '/content/dataset/images/train', '/content/dataset/labels/train')
mover(val_imgs, '/content/dataset/images/val', '/content/dataset/labels/val')
```

```python
import os
import shutil
from glob import glob
from sklearn.model_selection import train_test_split

# Diretórios de origem
image_root = '/content/dataset/datasets/images'
label_root = '/content/dataset/datasets/labels'

# Coletar imagens de todas as subpastas
images = glob(os.path.join(image_root, '**', '*.jpg'), recursive=True)

print(f'Total de imagens encontradas: {len(images)}')
if len(images) == 0:
    raise Exception("Nenhuma imagem encontrada! Verifique o caminho ou extensões.")

# Dividir em treino e validação (80/20)
train_imgs, val_imgs = train_test_split(images, test_size=0.2, random_state=42)

# Criar estrutura de diretórios
for path in [
    '/content/dataset/images/train',
    '/content/dataset/images/val',
    '/content/dataset/labels/train',
    '/content/dataset/labels/val'
]:
    os.makedirs(path, exist_ok=True)

# Função para mover imagens e seus labels
def move_images(image_list, split):
    for img_path in image_list:
        # Nome base do arquivo (sem pasta)
        filename = os.path.basename(img_path)
        name, _ = os.path.splitext(filename)

        # Caminho do label correspondente
        label_path = os.path.join(label_root, f'{name}.txt')

        # Destinos
        img_dst = f'/content/dataset/images/{split}/{filename}'
        lbl_dst = f'/content/dataset/labels/{split}/{name}.txt'

        # Copiar imagem
        shutil.copy(img_path, img_dst)

        # Copiar label se existir
        if os.path.exists(label_path):
            shutil.copy(label_path, lbl_dst)

# Mover arquivos
move_images(train_imgs, 'train')
move_images(val_imgs, 'val')

print('✅ Dataset organizado com sucesso!')
```

```python
# filtrar os arquivos de label
# Define as classes que você quer manter
target_classes = [0, 1]

# Caminhos para labels de treino e val
for split in ['train', 'val']:
    label_dir = f'/content/dataset/labels/{split}'
    label_files = glob(os.path.join(label_dir, '*.txt'))

    for label_file in label_files:
        new_lines = []
        with open(label_file, 'r') as f:
            lines = f.readlines()
            for line in lines:
                class_id = int(line.split()[0])
                if class_id in target_classes:
                    new_lines.append(line)

        # Substitui o conteúdo com apenas as classes desejadas
        if new_lines:
            with open(label_file, 'w') as f:
                f.writelines(new_lines)
        else:
            # Remove o label e sua imagem correspondente se não restou nenhuma anotação
            os.remove(label_file)
            img_file = label_file.replace('/labels/', '/images/').replace('.txt', '.jpg')
            if os.path.exists(img_file):
                os.remove(img_file)

print("✅ Labels filtrados. Apenas classes 0 e 1 foram mantidas.")
```
